```python
data = {'Название СК': [], 
         'Страховые премии тыс. руб.': [], 
         'Количество договоров страхования': [], 
         'Количество страховых случаев заявленных всего': [], 
         'Количество страховых случаев урегулированных': [], 
         'Количество страховых случаев из них отказов в выплате': [], 
         'Выплаты по договором страхования всего в тыс. руб.': []}

for i in name_list:
    try:
        df = pd.read_excel(f'companies_2018/{i}', sheet_name = 'Раздел 1')
        data['Название СК'].append(df.iat[12,3])
        data['Страховые премии тыс. руб.'].append(df.iat[67,2])
        data['Количество договоров страхования'].append(df.iat[67,3])
        data['Количество страховых случаев заявленных всего'].append(df.iat[67,5])
        data['Количество страховых случаев урегулированных'].append(df.iat[67,6])
        data['Количество страховых случаев из них отказов в выплате'].append(df.iat[67,7])
        data['Выплаты по договором страхования всего в тыс. руб.'].append(df.iat[67,8])
    except:
        pass

final_df = pd.DataFrame(data)
final_df   
```

```python
final_df.isna()
```

```python
final_df.where(~(final_df == '-'), other= None, inplace=True)
final_df.where(~(final_df == 0), other= None, inplace=True)
```

```python
final_df_without_NaN = final_df.dropna()
```

```python
final_df_without_NaN.to_excel('Показатели СК за целый 2018.xlsx')
```

```python
final_df_without_NaN
```

```python
Total = final_df_without_NaN['Страховые премии тыс. руб.'].sum()
share_df = final_df_without_NaN['Страховые премии тыс. руб.']/Total
share_df = {'Доля СК на рынке': share_df}
share_df = pd.DataFrame(share_df)
share_df
```

```python
DF = pd.merge(final_df_without_NaN, share_df, left_on='Название СК', right_on='Доля СК на рынке', how='left')
DF['Доля СК на рынке'] = DF['Страховые премии тыс. руб.']/Total
DF
```

```python
DF = DF[DF['Доля СК на рынке'] > 0.015]
DF
```
